```python

file = open('input.txt','r')

ran = []
ids = []

for line in file:
    line = line.strip()
    if '-' in line:
        ran.append(line)
    else:
        ids.append(line)

ids = ids[1:]
```

```python
seen = []
total = 0

#I add all the numbers of start/end of range
for r in ran:
    x = r.split('-')
    x[0] = int(x[0])
    x[1] = int(x[1])
    seen.append([x[0],x[1]])

#For each id, I check for each range if they are in between, and break at the firs true value
for id in ids:
    for r in seen:
        if int(id) >= r[0] and int(id) <= r[1]:
            total += 1
            break

total
```

```python

# I add all the ranges in a list
seen = []

for r in ran:
    x = r.split('-')
    x[0] = int(x[0])
    x[1] = int(x[1])
    seen.append([x[0],x[1]])


# I sort the ranges
seen.sort()

l = len(seen)
x = 1


# For each list, i check the next one and combine them if they overlap
while x < l:
    if int(seen[x][0]) <= int(seen[x-1][1]) and int(seen[x][0])>=int(seen[x-1][0]):
        if int(seen[x][1])>int(seen[x-1][1]):
            seen[x-1][1] = seen[x][1]
            seen.pop(x)
            x -= 1
        # I removed overlap that don't create correction
        else:
            seen.pop(x)
            x -= 1
        
    
    x += 1
    l = len(seen)


seen
```

```python
# As i only have now distinct ranges, I can only calculate the difference of each and add 1 (as range are inclusive)

total = 0
for i in seen:
    total += (i[1] - i[0])+1
    print(i, total)

print(total)
```
