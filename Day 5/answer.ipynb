```python

file = open('input.txt','r')

ran = []
ids = []

for line in file:
    line = line.strip()
    if '-' in line:
        ran.append(line)
    else:
        ids.append(line)

ids = ids[1:]
```

```python
ran = [
    '3-5',
    '10-14',
    '16-20',
    '12-18'
]

ids = ['1','5','8','11','17','32']
```

```python
seen = []
total = 0

#I add all the numbers of start/end of range
for r in ran:
    x = r.split('-')
    x[0] = int(x[0])
    x[1] = int(x[1])
    seen.append([x[0],x[1]])


for id in ids:
    for r in seen:
        if int(id) >= r[0] and int(id) <= r[1]:
            total += 1
            break

total
```
