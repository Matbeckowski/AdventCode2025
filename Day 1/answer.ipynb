```python


file = open('input.txt','r')

dir = []

for line in file:
    line = line.strip()
    dir.append(line)


print(len(dir))
```

```python
# Part one

dial = 50
nb_zero = 0

for i in dir:
    if i[:1] == 'L':
        val = -1
    else:
        val = 1

    for x in range(int(i[1:])):
        if dial == 99 and val == 1:
            dial = 0
        elif dial == 0 and val == -1:
            dial = 99
        else:
            dial += val
    
    if dial == 0:
        nb_zero += 1

        




print(nb_zero)
```
