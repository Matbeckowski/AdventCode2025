```python
file = [
'aaa: you hhh'
,'you: bbb ccc'
,'bbb: ddd eee'
,'ccc: ddd eee fff'
,'ddd: ggg'
,'eee: out'
,'fff: out'
,'ggg: out'
,'hhh: ccc fff iii'
,'iii: out'
]

grid = []

for line in file:
    line = line.strip()
    line = line.split(': ')
    grid.append(line)

grid
```

```python
paths = {}

for p in grid:
    paths[p[0]] = p[1].split(' ')


paths
```

```python

start = 'you'
seen = []
queue = [[start]]
x = 0

while queue:
    q = queue[0]
    mq = q[len(q)-1]

    if mq == 'out':
        seen.append(q)
        queue.pop(0)
    elif q not in seen:
        for i in paths[mq]:
            w = q.copy()
            w.append(i)
            queue.append(w)
        queue.pop(0)

print(len(seen))
```
