```python
# I create the list to stock the bad_ids
bad_id = []


for t in ids:
    # Check the range from the start to end IDs
    for i in range(int(t.split('-')[0]),int(t.split('-')[1])+1):
        txt = str(i)
        # I can skip all the numbers with an odd number of digits
        if len(txt)%2 == 1:
            pass
        else:
            # I split the string of the number in half, and check if both halves are equal
            mid = int(len(txt)/2)
            l = txt[:mid]
            r = txt[mid:]
            if l == r:
                bad_id.append(i)
```

```python
bad_id = []


for t in ids:
        # Check the range from the start to end IDs
    for i in range(int(t.split('-')[0]),int(t.split('-')[1])+1):
        txt = str(i)

        # I split the odd number by chunk of odd number (1,3,5,...) and split by the first occurence.
        # The number must have at least 2 digits
        # If these list only return '', then it means there is a pattern, otherwise it will return the series of digit that disrupt the pattern
        if len(txt)%2 == 1 and len(txt)>1:
            for j in range(1,int(len(txt)/2+1.5),2):
                l = [x for x in txt.split(txt[:j]) if x != '']
                if len(l)==0:
                    bad_id.append(i)
                    break
        else:
            # I split the even number into chuncks of 2 to half his len and check the split of the first occurence.
            # If these list only return '', then it means there is a pattern, otherwise it will return the series of digit that disrupt the pattern

            for j in range(1,int(len(txt)/2+1)):
                l = [x for x in txt.split(txt[:j]) if x != '']
                if len(l)==0:
                    bad_id.append(i)
                    break

print(sum(bad_id))
```
