```python


file = open('input.txt','r')

bank = []

for line in file:
    line = line.strip()
    bank.append(line)
```

Part ONE

```python
ans = 0

for b in bank:

    b = [int(x) for x in b]

    m = max(b[:-1])

    m_pos = b.index(m)

    m2 = max(b[m_pos+1:])

    ans += int(str(m)+str(m2))

ans
```
