```python


file = open('input.txt','r')

bank = []

for line in file:
    line = line.strip()
    bank.append(line)
```

Part ONE

```python

ans = 0

for b in bank:

    #transform string to integer
    b = [int(x) for x in b]

    #looking at the max value except the last digit, and find its position
    m = max(b[:-1])

    m_pos = b.index(m)

    #I find the max value from the remaining of the digits and concat them
    m2 = max(b[m_pos+1:])

    ans += int(str(m)+str(m2))

ans
```

Part Two

```python
ans = 0

for b in bank:

    b = [int(x) for x in b]

    
    length_bank = len(b)
    length_jolt = 12
    start = 0
    end = length_bank - length_jolt + 1
    jol = ''
    target = 12

    # Until my jolt is 12 digits long, I'm checking in the list the max value without the end of the list
    # I calculate how many numbers I still need to complete the 12 series, and I cut the remaining of the list so that always have enough digits to finish the series.
    while len(jol) < target:
        # find the max value and its position, and concat it the string result
        m = max(b[:end])
        m_pos = b.index(m)
        jol += str(m)
        # I take out the part left to the max value from the list, and update the remaining len of the jolt to find
        start = m_pos+1
        length_jolt -= 1
        # The list starts now from the highest previous value
        b = b[start:]
        # I cut the end to always have enough value to complete the string to 12 in case
        end = len(b) - (target-len(jol)) + 1
        #if what remains = nb of digit left to find, I just add them one by one
        if length_jolt == len(b):
            for i in b:
                jol += str(i)
  

    ans += int(jol)

print('answer :',ans)
```
