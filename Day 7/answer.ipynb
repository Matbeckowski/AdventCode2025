```python

file = open('input.txt','r')

grid = []

for line in file:
    line = line.strip()
    line = [x for x in line]
    grid.append(line)
```

```python
line = [
'.......S.......',
'...............',
'.......^.......',
'...............',
'......^.^......',
'...............',
'.....^.^.^.....',
'...............',
'....^.^...^....',
'...............',
'...^.^...^.^...',
'...............',
'..^...^.....^..',
'...............',
'.^.^.^.^.^...^.',
'...............'
]
```

```python
grid = []
for l in line:
    grid.append([x for x in l])

grid
```

```python
#Find the starting point
start = [0,grid[0].index('S')]
start

# I create the queue and the visited list where I will store the splitter I pass through
queue = [start]
visited = []

# The last row doesn't have any splitter, so I will remove any coordinate from the queue that reaches the last row directly, to avoid infinite loop
long = len(grid)-1

# Run a BFS algorythm
while queue:
    # If in the last row, I remove from the queue
    if queue[0][0] == long:
        queue.pop(0)
    # If i find a splitter that hasn't been visited yet, I add the left and right of the point to the queue and add the splitter in the visited list
    elif grid[queue[0][0]+1][queue[0][1]]  == '^' and [queue[0][0]+1,queue[0][1]] not in visited:
        queue.append([queue[0][0],queue[0][1]+1])
        queue.append([queue[0][0],queue[0][1]-1])
        visited.append([queue[0][0]+1,queue[0][1]])
        queue.pop(0)
    # Otherwise I change . -> | and I add continue going down by adding the next point to the queue
    else:
        grid[queue[0][0]+1][queue[0][1]] = '|'
        queue.append([queue[0][0]+1,queue[0][1]])
        queue.pop(0)


```
