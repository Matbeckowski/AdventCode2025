```python
start = [0,grid[0].index('S')]
start

queue = [start]
visited = []
splitter = 0

long = len(grid)-1


while queue:
    if queue[0][0] == long:
        queue.pop(0)
    elif grid[queue[0][0]+1][queue[0][1]]  == '^':
        splitter += 1
        queue.append([queue[0][0]+1,queue[0][1]+1])
        queue.append([queue[0][0]+1,queue[0][1]+1])
        queue.pop(0)
    else:
        grid[queue[0][0]+1][queue[0][1]] = '|'
        queue.append([queue[0][0]+1,queue[0][1]])
        queue.pop(0)


```
