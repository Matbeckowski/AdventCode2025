```python

file = open('input.txt','r')

grid = []

for line in file:
    line = line.strip()
    grid.append(line)
```

```python
grid = [
    '123 328  51  64',
    '45  64 387 23',
    '6  98 215 314'
    ,'* + * +'
]
```

```python

# I store the numbers in a list
rows = []

for line in grid:
    line = line.split(' ')
    line = [x for x in line if x != '']
    rows.append(line)

rows
```

```python
# I store the operations in a list
op = rows[-1:][0]
op

total = 0

# For each columns, I either add row by row or multiply row by row
for j in range(len(rows[0])):
    temp = 0
    for i in rows[:-1]:
        if op[j] == '+':
            temp += int(i[j])
        else:
            # if i don't put this line, then it will always * by 0
            if temp == 0:
                temp = 1
            temp *= int(i[j])
    total += temp

    
print(total)

```
