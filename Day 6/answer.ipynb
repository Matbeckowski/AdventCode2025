```python

file = open('input.txt','r')

grid = []

for line in file:
    line = line.strip()
    grid.append(line)
```

```python


rows = []

for line in grid:
    line = line.split(' ')
    line = [x for x in line if x != '']
    rows.append(line)

rows
```

```python
op = rows[-1:][0]
op

total = 0


for j in range(len(rows[0])):
    temp = 0
    print(j,'columns')
    for i in rows[:-1]:
        if op[j] == '+':
            temp += int(i[j])
        else:
            if temp == 0:
                temp = 1
            temp *= int(i[j])
    total += temp

    
print(total)

```
