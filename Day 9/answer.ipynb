```python
import math

file = open('input.txt','r')

grid = []

for line in file:
    line = line.strip()
    line = line.split(',')
    line = tuple([int(x) for x in line])
    grid.append(line)
```

```python
from itertools import combinations
import numpy as np

surface = {}

# I calculate the surface for each combination
# Surface of rectangle is equal to length * width
for r in list(combinations(grid,2)):
    if r[0][0] != r[1][0] and r[0][1]!=r[1][1]:
        a = abs(r[0][0]-r[1][0])+1
        b = abs(r[0][1]-r[1][1])+1
        c = a*b
        surface[r] = c
    
# Then I sort the dictionnary from the shortest to longest surface
keys = list(surface.keys())
values = list(surface.values())
sorted_value_index = np.argsort(values)[::-1]
sorted_dict = {keys[i]: values[i] for i in sorted_value_index}

list(sorted_dict.values())[0]
```

```python
lines = [
'7,1',
'11,1',
'11,7',
'9,7',
'9,5',
'2,5',
'2,3',
'7,3'
]

grid = []

for line in lines:
    line = line.split(',')
    line = [int(x) for x in line]
    line = tuple(line)
    grid.append(line)

grid.append(grid[0])
grid
```

```python
green = []
x = 0
for i in range(2,len(grid),1):
    a = grid[x:i]
    a0 = min(a[0][0],a[1][0])
    a1 = min(a[0][1],a[1][1])
    a2 = max(a[0][0],a[1][0])
    a3 = max(a[0][1],a[1][1])


    if a[0][0] == a[1][0]:
        a_green = [x for x in range(a1,a3+1)]
        a_green = list(zip([a[0][0]]*len(a_green),a_green))
        for ag in a_green:
            if ag not in green:
                green.append(ag)
    else:
        a_green = [x for x in range(a0,a2+1)]
        a_green = list(zip(a_green,[a[0][1]]*len(a_green)))
        for ag in a_green:
            if ag not in green:
                green.append(ag)
    x = i-1

green
```
