```python
from itertools import combinations
import numpy as np
import math

coord = []

for i in grid:
    coord.append(tuple(int(x) for x in i.split(',')))

coord

```

```python
d = {}

for c in list(combinations(coord, 2)):
    d[c] = math.dist(c[0],c[1])


keys = list(d.keys())
values = list(d.values())
sorted_value_index = np.argsort(values)
sorted_dict = {keys[i]: values[i] for i in sorted_value_index}

sorted_dict
```

```python
top = list(sorted_dict.keys())[:35]

top
```

```python
c = [[x for x in top[0]]]
c

for i in top[1:]:
    i = [x for x in i]
    for j in range(len(c)):
        if i[0] in c[j] and i[1] in c[j]:
            break
        elif i[0] in c[j]:
            c[j].append(i[1])
            break
        elif i[1] in c[j]:
            c[j].append(i[0])
            break
        elif j == len(c)-1:
            c.append(i)
            break


```

```python
l = [len(x) for x in c]
l.sort(reverse=True)
print(l[0]*l[1]*l[2])
```
