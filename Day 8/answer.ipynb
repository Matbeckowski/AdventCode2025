```python
file = open('input.txt','r')

grid = []

for line in file:
    line = line.strip()
    grid.append(line)
```

```python
from itertools import combinations
import numpy as np
import math

coord = []

for i in grid:
    coord.append(tuple(int(x) for x in i.split(',')))

```

```python
d = {}

# I store all the possible distances into a dictionnary
for c in list(combinations(coord, 2)):
    d[c] = math.dist(c[0],c[1])


# Then I sort the dictionnary from the shortest to longest distance
keys = list(d.keys())
values = list(d.values())
sorted_value_index = np.argsort(values)
sorted_dict = {keys[i]: values[i] for i in sorted_value_index}

```

```python
# I keep only the X nb of connections we want to look at

connections = 1000

top = list(sorted_dict.keys())[:connections]
```

```python
# I initiate the first circuit
c = [[x for x in top[0]]]

# For each combination, I check match in every circuit
for i in top[1:]:
    t = [x for x in i]
    for j in c:
        # Check if any match
        t0 = [x for x in j if x in t]
        if len(t0)>0:
            # I combine the list and keep the unique items
            t = list(set(j+t))
            # I remove the now duplicate
            c.pop(c.index(j))
            
    # I append the new list of tuples
    c.append(t)
    




```

```python
l = [len(x) for x in c]
l.sort(reverse=True)
print(l[0],l[1],l[2])
print(l[0]*l[1]*l[2])
```
